# Análisis temporal de la relación entre temperatura e irradiancia solar

La temperatura y la irradiancia solar son variables fundamentales en el estudio del comportamiento atmosférico. Estas variables presentan una relación directa por diversos factores físicos y temporales.

En este análisis se estudia la temperatura e irradiancia solar de Khulna (Ciudad en Bangladés) a lo largo del tiempo, así como identificar patrones, desfases o posibles inconsistencias en su comportamiento.

**Objetivo**: Analizar la relación entre la irradiancia solar y la temperatura en función del tiempo, identificando patrones diarios y mensuales, así como posibles desfases térmicos y las variaciones en dicha relación.

## Análisis gráfco del comportamiento de Temperatura e irradiancia por hora
Para analizar el comportamiento diario de la temperatura y la irradiancia, se calcula el promedio de ambas variables por hora.

```python
df_hr = df.groupby("Hour")[["Temperature","Irradiance"]].mean().reset_index()
df_hr.head()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(df_hr["Hour"], df_hr["Temperature"], label="Temperature")
ax1.set_ylabel("Temperature")

ax2 = ax1.twinx()
ax2.plot(df_hr["Hour"], df_hr["Irradiance"], color = 'r', linestyle="--", label="Irradiance")
ax2.set_ylabel("Irradiance")

plt.title("Temperature & Irradiance by Hour")

fig.legend()
plt.show()
```

Se puede observar que ambas variables presentan un comportamiento similar, sin embargo, la temperatura alcanza su máximo aproximadamente 2 a 3 horas despues que la irradiancia solar, esto evidencia un desfase térmico asociado a la acumulación progresiva de calor en el ambiente.

## Análisis gráfco del comportamiento de Temperatura e irradiancia por mes

```python
df_month = df.groupby("Month")[["Temperature","Irradiance"]].mean().reset_index()
df_month.head()
```

```python
fig, ax3 = plt.subplots()

ax3.plot(df_month["Month"], df_month["Temperature"], label="Temperature")
ax3.set_ylabel("Temperature")

ax4 = ax3.twinx()
ax4.plot(df_month["Month"], df_month["Irradiance"], color = 'r', linestyle="--", label="Irradiance")
ax4.set_ylabel("Irradiance")

plt.title("Temperature & Irradiance by Month")

fig.legend()
plt.show()
```

A diferencia del comportamiento por hora, la relación entre temperatura e irradiancia varia significativamente a lo largo de los meses, y con el objetivo de analizar si la relación temporal entre temperatura e irradiancia cambia a lo largo del año, se comparó el comportamiento horario promedio de ambas variables en los meses de abril y julio.

```python
df_abril = df[df["Month"] == 4]
```

```python
df_abril_hr = df_abril.groupby("Hour")[["Temperature", "Irradiance"]].mean().reset_index()
df_abril_hr.head()
```

```python
fig, ax5 = plt.subplots()

ax5.plot(df_abril_hr["Hour"], df_abril_hr["Temperature"], label = "Temperature")
ax5.set_ylabel("Temperature")

ax6 = ax5.twinx()
ax6.plot(df_abril_hr["Hour"], df_abril_hr["Irradiance"], color = 'r', linestyle="--", label = "Irradiance")
ax6.set_ylabel("Irradiance")

plt.title("(April)Temperature & Irradiance by Hour")

fig.legend()
plt.show()
```

```python
dF_jul = df[df["Month"] == 7]
```

```python
df_jul_hr = dF_jul.groupby("Hour")[["Temperature", "Irradiance"]].mean().reset_index()
df_jul_hr.head()
```

```python
fig, ax7 = plt.subplots()

ax7.plot(df_jul_hr["Hour"], df_jul_hr["Temperature"], label="Temperature")
ax7.set_ylabel("Temperature")

ax8 = ax7.twinx()
ax8.plot(df_jul_hr["Hour"], df_jul_hr["Irradiance"], color = 'r', linestyle="--", label="Irradiance")
ax8.set_ylabel("Irradiance")

plt.title("(July)Temperature & Irradiance by Hour")

fig.legend()
plt.show()
```

Aunque la relación entre temperatura e irradiancia presenta variaciones mensuales en su magnitud, el comportamiento horario mantiene una dinámica similar entre distintos meses, conservando el desfase térmico de 2 a 3 horas observado previamente.

En abril se observan valores máximos mayores tanto de irradiancia como de temperatura respecto a julio, lo que indica una diferencia significativa en intensidad entre ambos meses.

Por lo tanto, las diferencias observadas entre Abril y Julio no corresponden a un cambio en la dinámica temporal de las variables, sino a las variaciones de magnitud de irradiancia y temperatura entre ambos meses.

## Variabilidad de la irradiancia entre abril y julio
A continuación se compara la dispersión de los valores de irradiancia entre ambos meses.

```python
df_box = df[df["Month"].isin([4, 7])]
```

```python
sns.boxplot(x = 'Month', y = 'Irradiance', color= 'cyan', data = df_box)
plt.title('Month vs Irradiance')
plt.xlabel('Month (4 = April, 7 =July)')
plt.ylabel('Irradiance')
plt.show()
```

Abril presenta valores de irradiancia más altos y una mayor dispersión respecto a julio, lo que coincide con las diferencias de magnitud observadas previamente en el análisis horario.

## Conclusiones

El análisis realizado permitio identificar una correlación positiva moderada entre la irradiancia solar y la temperatura, indicando una relación directa, aunque no estrictamente lineal.

A nivel horario, se observo un comportamiento temporal similar entre ambas variables, identificando un desfase térmico aproximadamente de 2 a 3 horas, donde la temperatura alcanzo su máximo despues del pico de irradiancia solar. En el análisis mensual se encontraron diferencias significativas en magnitud entre meses como abril y julio, sin embargo, la dinámica temporal de las variables se mantuvo consistente.

Finalmente, se concluye que, a pesar de que existe factores atmosféricos y ambientales que no se incluyen en el dataset, la irradiancia solar influye directamente en el comportamiento térmico observado.
